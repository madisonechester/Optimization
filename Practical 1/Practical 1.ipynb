```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D  

# defining the surface
X1 = np.linspace(-2, 2, 400)
X2 = np.linspace(-2, 2, 400)
X1, X2 = np.meshgrid(X1, X2)

# defining the functions
f_A = -X1**2 - X2**2
f_B = X1**2 - X2**2
f_C = X1**2

# 3D surface plots and contour plots
fig = plt.figure(figsize=(15, 5))

ax1 = fig.add_subplot(131, projection='3d')
ax1.plot_surface(X1, X2, f_A, cmap=cm.coolwarm, linewidth=0, antialiased=False)
ax1.contour(X1, X2, f_A, zdir='z', offset=0)
ax1.set_title('Contour Plot of f_A(x)')

ax2 = fig.add_subplot(132, projection='3d')
ax2.plot_surface(X1, X2, f_B, cmap=cm.coolwarm, linewidth=0, antialiased=False)
ax2.contour(X1, X2, f_B, zdir='z', offset=0)
ax2.set_title('Contour Plot of f_B(x)')

ax3 = fig.add_subplot(133, projection='3d')
ax3.plot_surface(X1, X2, f_C, cmap=cm.coolwarm, linewidth=0, antialiased=False)
ax3.contour(X1, X2, f_C, zdir='z', offset=0)
ax3.set_title('Contour Plot of f_C(x)')

plt.tight_layout()
plt.show()
```

```python
import numpy as np

# defining the Hessian matrices 
Hessian_A = np.array([[-2, 0], [0, -2]])
Hessian_B = np.array([[2, 0], [0, -2]])
Hessian_C = np.array([[2, 0], [0, 0]])

# computing the eigenvalues 
eigenvalues_A = np.linalg.eigvals(Hessian_A)
eigenvalues_B = np.linalg.eigvals(Hessian_B)
eigenvalues_C = np.linalg.eigvals(Hessian_C)

# analyzing the eigenvalues
def analyze_eigenvalues(eigenvalues):
    if all(eigval > 0 for eigval in eigenvalues):
        return "Minimum - Convex"
    elif all(eigval < 0 for eigval in eigenvalues):
        return "Maximum - Concave"
    else:
        return "Neither - Saddle"

# analyzing the eigenvalues for each function
result_A = analyze_eigenvalues(eigenvalues_A)
result_B = analyze_eigenvalues(eigenvalues_B)
result_C = analyze_eigenvalues(eigenvalues_C)

# printing the results
print("For f_A(x):", result_A)
print("For f_B(x):", result_B)
print("For f_C(x):", result_C)
```

### 2.2 - A two dimensional function with multiple minima

#### 2.2.1

```python
import warnings
warnings.filterwarnings("ignore")
import numpy as np
import matplotlib.pyplot as plt

# defining the surface
X1 = np.arange(-2, 2, 0.1)
X2 = np.arange(-1, 1, 0.1)
X1, X2 = np.meshgrid(X1, X2)

# defining the function
def f(X1, X2):
    return X1**2 * (4 - 2.1 * X1**2 + (1/3) * X1**4) + X1 * X2 + X2**2 * (-4 + 4 * X2**2)

# calculating the values of f(x1, x2)
Z = f(X1, X2)

# 3D surface plot and contour plot 
fig = plt.figure(figsize = (8,6))
ax = fig.gca(projection='3d')
ax.plot_surface(X1, X2, Z, rstride=1, cstride=1, cmap=cm.coolwarm, linewidth=0, antialiased=False)
ax.contour(X1,X2,Z,30,zdir='z',offset=0)

plt.show()
```

```python
import numpy as np

# defining the surface
x1_range = np.arange(-2, 2, 0.005)
x2_range = np.arange(-1, 1, 0.005)
x1, x2 = np.meshgrid(x1_range, x2_range)

# defining the gradient components 
df_dx1 = 2*x1**5 - 8.4*x1**3 + 8*x1 + x2
df_dx2 = x1 - 8*x2 + 16*x2**3

# calculating the squared norm of the gradient
gradient_norm_squared = df_dx1**2 + df_dx2**2

# storing in a matrix
result_matrix = np.column_stack((x1.flatten(), x2.flatten(), gradient_norm_squared.flatten()))
np.savetxt("gradient_norm_squared.csv", result_matrix, delimiter=",", header="x1,x2,gradient_norm_squared")
```

```python
import numpy as np

# loading the gradient_norm_squared data
data = np.loadtxt("gradient_norm_squared.csv", delimiter=",", skiprows=1)
x1_values = data[:, 0]
x2_values = data[:, 1]
gradient_norm_squared_values = data[:, 2]

# reshaping to match the mesh grid shape
gradient_norm_squared_grid = gradient_norm_squared_values.reshape(len(x2_range), len(x1_range))

# initializing a list to store candidate points and their gradient norm squared values
values_at_candidate_points = []

# iterating through the grid
for i in range(1, len(x1_range) - 1):
    for j in range(1, len(x2_range) - 1):
        # the gradient norm squared value at each point
        current_point_value = gradient_norm_squared_grid[j, i]
        
        # gradient norm squared value of the 8 neighbors 
        top_left_neighbor = gradient_norm_squared_grid[j-1, i-1]
        top_center_neighbor = gradient_norm_squared_grid[j-1, i]
        top_right_neighbor = gradient_norm_squared_grid[j-1, i+1]
        left_neighbor = gradient_norm_squared_grid[j, i-1]
        right_neighbor = gradient_norm_squared_grid[j, i+1]
        bottom_left_neighbor = gradient_norm_squared_grid[j+1, i-1]
        bottom_center_neighbor = gradient_norm_squared_grid[j+1, i]
        bottom_right_neighbor = gradient_norm_squared_grid[j+1, i+1]
        
        # comparing the gradient norm squared of the point and its neighbors 
        if (current_point_value < top_left_neighbor and
            current_point_value < top_center_neighbor and
            current_point_value < top_right_neighbor and
            current_point_value < left_neighbor and
            current_point_value < right_neighbor and
            current_point_value < bottom_left_neighbor and
            current_point_value < bottom_center_neighbor and
            current_point_value < bottom_right_neighbor):
            x1, x2 = x1_range[i], x2_range[j]
            grad_norm_sq_value = current_point_value
            values_at_candidate_points.append((x1, x2, grad_norm_sq_value))
```

```python
# printing the point and the gradient norm squared value 
print("Candidate Points:")
for point in values_at_candidate_points:
    x1, x2, grad_norm_sq_value = point
    print("Point: (x1 =", x1, ", x2 =", x2, ")")
    print("||f(x)||^2 Value:", grad_norm_sq_value)
    print("---")
```

```python
import numpy as np
import sympy as sp

# defining the function f(x1, x2)
x1, x2 = sp.symbols('x1 x2')
f = x1**2 * (4 - 2.1 * x1**2 + (1/3) * x1**4) + x1 * x2 + x2**2 * (-4 + 4 * x2**2)

# computing the Hessian matrix
Hessian_matrix = sp.hessian(f, (x1, x2))

# values of candidate points
candidate_points = [(x1_value, x2_value) for x1_value, x2_value, _ in values_at_candidate_points]

# initializing a list to store critical point types
point_types = []

# computing the eigenvalues of the Hessian matrix
for point in candidate_points:
    Hessian_numeric = np.array(Hessian_matrix.subs({x1: point[0], x2: point[1]}), dtype=float)
    eigenvalues = np.linalg.eigvals(Hessian_numeric)
    
    if all(eigval > 0 for eigval in eigenvalues):
        point_types.append("Minimum")
    elif all(eigval < 0 for eigval in eigenvalues):
        point_types.append("Maximum")
    else:
        point_types.append("Saddle")

# printing the critical point types
for i, point in enumerate(candidate_points):
    print(f"Point {point}: {point_types[i]}")
```
